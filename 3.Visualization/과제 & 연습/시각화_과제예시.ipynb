```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel("../dataSet/auto-mpg.xlsx")
df1
```

```python
sns.relplot(data = df1, x = 'weight', y = 'mpg', hue = "origin", palette=['skyblue', 'limegreen', 'red'])
```

```python
fig, ax = plt.subplots()
for key, grp in df1.groupby(['origin']):
    ax.scatter(x='weight', y='mpg', data=grp, label=key)

ax.legend()
plt.show()
```

```python
sns.scatterplot(data = df1,x = "displacement",y = "mpg")
```

```python
df1[["displacement","mpg"]].plot(kind = "scatter", x = "displacement", y ="mpg")
```

```python
plt.scatter(data = df1, x = "displacement", y = "mpg")
```

```python
sns.lineplot(data = df1, x = "model year", y = "mpg", hue = "origin", palette=['red', 'green', 'blue'])
```

```python
fig, ax = plt.subplots()
for key, grp in df1.groupby(['origin']):
    ax.plot(grp['model year'], grp['mpg'], label=key)
ax.legend()
plt.show()
```
