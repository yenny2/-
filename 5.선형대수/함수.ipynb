#### 역함수
- 어떤 함수의 입력/출력 관계와 정반대의 입출력 관계를 갖는 함수
- 원래의 함수에서 x축과 y축이 바뀐 것이므로 y=x가 나타내는 직선을 대칭축으로 대칭인 함수의 그래프

```python
def f1(x):
    return x ** 2
def f1inv(x):
    return np.sqrt(x)     # x 자승의 역함수는 루트!

x = np.linspace(0,3,300)
plt.figure(figsize = (10,6))
plt.plot(x,f1(x),"r-", label="함수 $f(x) = x^2")   # x ** 2 = x^
plt.plot(x,f1inv(x),"b-", label="역함수 $f^{-1}(x) = \sqrt{x}$")
plt.plot(x,x,"g--")
plt.axis('equal')
plt.xlim(0,2)
plt.ylim(0,2)
plt.legend()
plt.title('역함수의 그래프')
plt.show
```

### 다음 함수의 역함수를 구하고 그래프로 표현

- 원래 함수의 x와 y를 바꿔줍니다
- 바뀐 함수를 y에 대해서 푼 후, x로 정리합니다.

```python
def f(x):
    return (x ** 2 + 1)

def finv(x):
    return(np.sqrt(x-1))

x = np.linspace(0,3,300)
plt.figure(figsize = (10,6))
plt.plot(x,f(x),"r-", label="함수 $f(x) = x^2+1")   # x ** 2 = x^
plt.plot(x,finv(x),"b-", label="역함수 $f^{-1}(x) = \sqrt{x-1}$")
plt.plot(x,x,"g--")
plt.axis('equal')
plt.xlim(0,2)
plt.ylim(0,2)
plt.legend()
plt.title('역함수의 그래프')
plt.show
```

```python
# ReLU : 딥 러닝에서 가장 많이 사용되는 활성화 함수

xx = np.linspace(-10,10,100)
plt.figure(figsize = (10,6))
plt.plot(xx, np.maximum(xx,0))
plt.title('max(x,0) 또는 ReLU')
plt.xlabel("$x$")
plt.ylabel("$ReLU(x)$")
plt.show()


# $ReLU(x) = max(0,x)$
# 이 함수는 입력 값이 0보다 작을 경우 0을 출력하고, 입력 값이 0보다 큰 경우 입력 값을 그대로 출력
```

#### 지수함수
- 정해진 어떤 값을 입력값으로 거듭제곱하여 출력을 구하는 함수
- 거듭제곱을 할 때 아래에 있는 수를 밑이라고 하며 밑을 오일러 수 e(약 2.718)로 하여 거듭제곱하는 함수를 지수함수라고 한다.
- $y=e^x$, y=exp(x) = exp x 로 표시
- 로그함수의 밑으로 자연로그 함수(In)을 사용하는데 이때 밑이 오일러 상수 e이다.

- 수식 쓸 때 $ 표시

```python
import math
print(math.e)
```

```python
# 로그 함수의 밑이 e인 경우

x = 10
y = math.log(x)
y
```

```python
# 로그 함수의 밑이 다른 수인 경우

x = 10
base = 10
y = math.log(x,base)
y
```

```python
print(np.exp(-10))   #  e의 -10 제곱값을 계산하고 출력
print(np.exp(-1))
print(np.exp(-0.1))
print(np.exp(0))
print(np.exp(0.1))
print(np.exp(1))
print(np.exp(10))
```

#### Q2. mt1 행렬에 대하여 다음을 수행하세요.

1. 역행렬 mt2를 구하고
2. mt1과 mt2를 내적하여 단위 행렬을 출력

```python
np.random.seed(0)
mt1 = np.random.randint(1,4,size = (3,3))
mt1

# 1.역행렬 mt2를 구하고

mt2 = np.linalg.inv(mt1) 
print(mt2)

# 2. mt1과 mt2를 내적하여 단위 행렬을 출력

mt1 @ mt2
```
