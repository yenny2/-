```python
def get_category(x):
    cat = ''
    if x <= 10:
        cat = 'Child'
    elif x <= 30:
        cat = 'Young Adult'
    elif x <=60:
        cat = 'Adult'
    else:
        'Elderly'
    return cat

df['age_class'] = df.age.apply(lambda x:get_category(x))             # 전체 행 or 열에 함수 적용: apply(함수, axis=0열 or 1행)
df.loc[:,['age','age_class']].head()
```

### 정규화
* 각 변수의 숫자 데이터의 상대적 크기 차이 때문에 ml 분석 결과가 달라질 수 있음. (a변수 1 ~ 1000, b변수 0 ~ 1)
* 숫자 데이터의 상대적 크기 차이를 제거할 필요가 있으며 각열(변수)에 속하는 데이터 값을 동일한 크기 기분으로 나누어 정규화 함.
* 정규화 결과 데이터의 범위는 0 ~ 1 또는 -1 ~ 1(음수값이 있는 경우). 
* 각 열의 값 / 최대값 or (각 열의 값 - 최소값) / (해당 열의 최대값 - 최소값)

#### 전처리, 모델 생성 및 평가

1. 전처리
    - 데이터셋을 로드하고, 컬럼명을 지정합니다.
    - horsepower 열의 결측값인 ?을 NaN으로 변경하고, 이를 포함하는 행을 삭제합니다.
    - car name 열을 삭제합니다.
    - 범주형 데이터인 origin을 one-hot 인코딩합니다.
    - 데이터를 입력 변수 X(독립변수)와 출력 변수 y(종속변수)로 분할합니다.
    - 입력 변수 X를 정규화합니다. (스케일이 다르면 정규화해줘야 함)<br>
<br>    
2. 모델 학습

    - 학습 데이터셋과 테스트 데이터셋으로 데이터를 분할합니다.
        - 학습 데이터셋으로 평가하면 안됨. 
        우리가 추구하는 것은 일반적으로 잘 작동하는 모델을 원하기에 테스트할 때는 학습한 데이터는 제외하고 새로운 데이터를 
    - 선형 회귀 모델을 생성하고, 학습 데이터셋으로 모델을 학습합니다.<br>
<br>    
3. 모델 성능 평가
    - 테스트 데이터셋으로 모델을 예측하고, R^2 점수를 계산하여 모델의 성능을 평가합니다.
    - mpg를 예측하는 선형 회귀 모델의 성능 평가
      - R^
          우리가 개발한 모델이 예측성능이 얼마나 되는지 평가
      - MSE (Mean Squared Error)와 RMSE (Root Mean Squared Error)
        - MSE는 예측값과 실제값의 차이를 제곱한 값들의 평균이고, RMSE는 MSE의 제곱근입니다.
        

※ R^2는 결정 계수(Coefficient of Determination)를 의미합니다. 결정 계수는 회귀 분석에서 독립 변수(X)가 종속 변수(Y)에 대해 얼마나 잘 설명하는지를 나타내는 지표로, 0에서 1사이의 값을 가집니다.

R^2의 값이 1에 가까울수록 회귀 모델이 데이터를 더 잘 설명한다는 의미입니다. 반대로 R^2의 값이 0에 가까울수록 모델이 데이터를 설명하는 능력이 낮다는 것을 의미합니다. R^2의 값이 음수일 수도 있는데, 이는 모델이 데이터보다 더 나쁘게 예측한 경우입니다.

R^2는 다른 성능 평가 지표와 함께 모델의 성능을 종합적으로 평가할 수 있습니다. 그러나 R^2는 독립 변수와 종속 변수 간의 관계를 단순화하므로, 모델의 성능 평가만으로는 충분하지 않을 수 있습니다. 따라서 다른 성능 평가 지표와 함께 사용하여 모델의 성능을 정확히 평가하는 것이 좋습니다.

```python
import pandas as pd

# 데이터셋 url
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"

# 컬럼명 지정
columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', \
           'origin', 'car name']

# 데이터 가져오기
df = pd.read_csv(url, names = columns, delim_whitespace = True)   # delim_whitespace = True 공백은 삭제
df.head()
```

```python
# 전처리
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# 데이터셋 url
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"

# 컬럼명 지정
columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', \
           'origin', 'car name']

# 데이터 가져오기
df = pd.read_csv(url, names = columns, delim_whitespace = True)   # delim_whitespace = True 공백은 삭제

# step1. horsepower 열의 결측값인 ?을 NaN으로 변경하고, 이를 포함하는 행을 삭제합니다.
df = df.replace('?', np.nan)
df = df.dropna()
# df['horsepower'].replace('?', np.nan, inplace=True) 
# df.dropna(subset=['horsepower'], axis=0, inplace=True)     # 난값 삭제

# step2. 불필요한 컬럼 삭제 (차 이름은 차 연비에 영향을 주지 않으니까 삭제)
df = df.drop('car name', axis=1) # 열인 카네임 삭제


# step3. 범주형 데이터인 'origin'을 one-hot 인코딩(숫자로 바꿔주는 것.)
df = pd.get_dummies(df, columns=['origin'])

# step4. 데이터 분할(독립변수, 종속변수)
x = df.drop('mpg',axis=1) # 종속변수만을 드랍해주면 독립변수가 남으니까)
y = df['mpg']

# step5. 스케일이 다른 것을 정규화(StandardScaler, 평균 0, 표준편차 1인 분포로 정규화하는 것)
# 위에서는 최대1 최소0으로 하는 정규화 MinMaxScaler() 했었음
    
scaler = StandardScaler()
x = scaler.fit_transform(x)             # fit_transform() 메서드를 이용해 x를 표준화해줌

# 결과확인

print(x[:5])
print(y[:5])
```
