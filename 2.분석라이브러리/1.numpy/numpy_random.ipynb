- id : 1번 ~ 1000번 일련번호
- gender : 0,1 정수 난수 생성
- age : 10 ~ 80사이 정수 난수를 생성
- region : 1 ~ 10 사이 정수 난수 생성
- prod_22 : 제품코드 11111 ~ 99999
- prod_23 : 제품코드 11111 ~ 99999
- price_22 : 1000 ~ 50000 사이 난수 생성
- price_23 : 1000 ~ 50000 사이 난수 생성
- qty_22 : 1 ~ 100 사이 정수 난수 생성하여('19년 구매 수량)
- qty_23 : 1 ~ 100 사이 정수 난수 생성하여('20년 구매 수량)
- time_22 : 01 ~ 24 사이 정수 난수 생성('19년에 가장 자주 구매한 시간대)
- time_23 : 01 ~ 24 사이 정수 난수 생성('20년에 가장 자주 구매한 시간대)
- amount_22: price_22 * qty_22
- amount_23: price_23 * qty_23
- sales : 22년 대비 23년 구매금액이 증가면 1, 감소면 0

```python
# 우선 필요한 데이터 프레임 틀을 만들기
ar = np.zeros((1000,15),dtype = int)

# id 1 ~ 1000
ar[:,0] = np.arange(1,1001)

# gender : 0,1 정수 난수 생성
ar[:,1] = np.random.choice([0,1],1000)

# age : 10 ~ 80사이 정수 난수를 생성
ar[:,2] = np.random.randint(10,81,1000)

# region : 1 ~ 10 사이 정수 난수 생성
ar[:,3] = np.random.randint(1,11,1000)

# prod_22 : 제품코드 11111 ~ 99999
ar[:,4] = np.random.randint(11111,100000,1000)
ar[:,5] = np.random.randint(11111,100000,1000)

# price_22 : 1000 ~ 50000 사이 난수 생성
ar[:,6] = np.random.randint(1,51,1000) * 1000
ar[:,7] = np.random.randint(1,51,1000) * 1000

# qty_22 : 1 ~ 100 사이 정수 난수 생성하여('19년 구매 수량)
# qty_23 : 1 ~ 100 사이 정수 난수 생성하여('20년 구매 수량)
ar[:,8] = np.random.randint(1,101,1000)
ar[:,9] = np.random.randint(1,101,1000)

# time_22 : 01 ~ 24 사이 정수 난수 생성('19년에 가장 자주 구매한 시간대)
# time_23 : 01 ~ 24 사이 정수 난수 생성('20년에 가장 자주 구매한 시간대)
ar[:,10] = np.random.randint(1,24,1000)
ar[:,11] = np.random.randint(1,24,1000)

# amount_22: price_22 * qty_22
# amount_23: price_23 * qty_23
ar[:,12] = ar[:,6] * ar[:,8]
ar[:,13] = ar[:,7] * ar[:,9]

# sales : 22년 대비 23년 구매금액이 증가면 1, 감소면 0
ar[:,14] = np.where(((ar[:,13] - ar[:,12]) > 0), 1, 0)

columns = ['id', 'gender', 'age', 'region', 'prod_22', 'prod_23', 'price_22', 'price_23', 'qty_22', 'qty_23', 'time_22', 'time_23', 'amount_22', 'amount_23', 'sales']
df = pd.DataFrame(ar,columns = columns)
df.head()


```
