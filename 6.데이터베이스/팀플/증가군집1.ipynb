```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:\hmkd1\m5_ml\머신러닝03_평가\dataset/증가군집1.csv", encoding="utf-8")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc("font", family="Malgun Gothic")
```

```python
df.연령대.value_counts()
```

```python
dfy=df.groupby('연령대').agg({"고객번호":"count"}).reset_index()
dfy
```

```python
dft=df.groupby('연령대').agg({"고객번호":"count"}).reset_index()
rc('font',family='Malgun Gothic')  # 한글 폰트 설정
plt.pie(dft["고객번호"], labels=dft["연령대"], autopct='%.1f%%')
plt.show()
ax = sns.histplot(df.연령대)

# 숫자 표시
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), 
                textcoords='offset points')
```

for p in ax.patches: ax.patches는 그래프의 모든 막대를 나타내는 패치(patch) 객체의 리스트입니다. for 루프를 통해 각 막대에 대한 작업을 수행합니다.

ax.annotate(...): annotate() 함수를 사용하여 숫자를 막대 위에 표시합니다.

f'{int(p.get_height())}': get_height() 메서드를 통해 현재 막대의 높이를 얻습니다. 이를 정수로 변환하여 문자열에 포맷팅합니다. 따라서 막대 위에 표시되는 숫자는 해당 막대의 높이입니다.

(p.get_x() + p.get_width() / 2., p.get_height()): 숫자를 표시할 위치입니다. get_x()와 get_width() 메서드를 통해 막대의 중앙 위치(x좌표)를 계산하고, get_height() 메서드를 통해 막대의 높이(y좌표)를 얻습니다.

ha='center', va='center': 텍스트의 가로 정렬(horizontal alignment)과 세로 정렬(vertical alignment)을 설정합니다. 'center'로 설정하여 가운데 정렬합니다.

xytext=(0, 5), textcoords='offset points': 텍스트의 위치 조정에 관한 옵션입니다. xytext는 텍스트 위치의 상대적 이동을 나타내며, (0, 5)는 막대 위에서 5 포인트 위로 이동하도록 설정합니다. textcoords는 xytext의 좌표 시스템을 나타냅니다.
