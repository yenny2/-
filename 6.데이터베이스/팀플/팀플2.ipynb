```python
# 원핫인코딩

df_14_1 = pd.get_dummies(df_14_1)
```

```python
df_14_1['감소고객'] = df_14_2['14-2분기']-df_14_1['14-1분기']
df_14_1.head(3)
```

```python
df_14_1.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# int로 데이터형식 바꾸기

df_14_1.iloc[:, -12:] = df_14_1.iloc[:, -12:].astype('int64')
```

```python
for i in range(len(df_14_1)):
    if df_14_1.loc[i, '감소고객'] > 0:
        df_14_1.loc[i, '감소'] = 0
    else:
        df_14_1.loc[i, '감소'] = 1
df_14_1
```

```python
df_14_1.drop('감소고객', axis=1, inplace = True)
```

```python
x = df_14_1.iloc[:,:-1]
y = df_14_1.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# 랜덤 포레스트 학습 및 별도의 테스트 세트로 예측 성능 평가
rf_clf = RandomForestClassifier(random_state=0, max_depth=8)
rf_clf.fit(x_train, y_train)
pred = rf_clf.predict(x_test)
accuracy = accuracy_score(y_test, pred)
print('랜덤 포레스트 정확도: {0: .4f}'.format(accuracy))
```

```python
# 그리드서치

from sklearn.model_selection import GridSearchCV

# 램덤포레스트 학습 데이터와 테스트 데이터 분리
rf = RandomForestClassifier()

# 파라미터를 딕셔너리 형태로 설정
param = {'max_depth':[8,9,10,11], 'min_samples_split':[5,6,7]}


# param_grid의 하이퍼 파라미터를 3개의 train, test set fold로 나눠 테스트 수행 설정
grid_rf = GridSearchCV(rf, param_grid=param, cv=3, refit=True)

# 학습 데이터로 param_grid의 하이퍼 파라미터를 순차적으로 학습/평가
grid_rf.fit(x_train, y_train)

# GridSearchCV 결과를 추출해 DataFrame으로 변환
scores_df = pd.DataFrame(grid_rf.cv_results_)
scores_df[['params','mean_test_score','rank_test_score',
          'split0_test_score', 'split1_test_score', 'split2_test_score']]
```
